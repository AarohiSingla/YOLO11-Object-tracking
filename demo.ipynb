#### OpenCV and YOLO11 to run object tracking on video frames. 

```python
import cv2
import time
from ultralytics import YOLO

# Load the YOLO11 model
model = YOLO("yolo11m.pt")

# Open the video file
video_path = "test_videos/2.mp4"
cap = cv2.VideoCapture(video_path)

# Initialize variables for FPS calculation
frame_count = 0
start_time = time.time()

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Start timing the frame processing
        frame_start_time = time.time()
        
        # Run YOLO11 tracking on the frame, persisting tracks between frames
        results = model.track(frame, classes=0, persist=True, save=True, tracker="bytetrack.yaml")
        
        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        
        # Calculate the time taken to process the frame
        frame_process_time = time.time() - frame_start_time
        
        # Update frame count
        frame_count += 1
        
        # Calculate FPS
        elapsed_time = time.time() - start_time
        if elapsed_time > 0:
            fps = frame_count / elapsed_time
        else:
            fps = 0
        
        # Display FPS and processing time on the frame
        cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(annotated_frame, f"Processing Time: {frame_process_time:.4f} s", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        # Display the annotated frame
        cv2.imshow("YOLO11 Tracking", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
